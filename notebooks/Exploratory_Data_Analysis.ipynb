```python
import pandas as pd

# Ruta al archivo de datos procesados
processed_data_path = "../data/processed/test_FD001_processed.csv"

# Cargar los datos
df = pd.read_csv(processed_data_path)
df.head()
```

```python
# DescripciÃ³n estadÃ­stica bÃ¡sica
df.describe()

# Visualizar la distribuciÃ³n de las caracterÃ­sticas
import seaborn as sns
import matplotlib.pyplot as plt

# Histograma de todas las columnas
df.hist(bins=50, figsize=(20,15))
plt.show()

# Matriz de correlaciÃ³n
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Seleccionar algunas columnas de interÃ©s para visualizar
columns_of_interest = [0, 1, 2, 3, 4]  # Cambia los nÃºmeros a las columnas que te interesen
for col in columns_of_interest:
    plt.plot(df.index, df.iloc[:, col], label=f'Feature {col}')

plt.xlabel("Cycle (Index)")
plt.ylabel("Sensor Readings")
plt.title("Sensor Readings Over Cycles")
plt.legend()
plt.show()
```

```python
# Supongamos que cada secuencia tiene N ciclos
N = 100  # NÃºmero de ciclos (esto es un ejemplo, ajÃºstalo segÃºn tus datos)

# Crear una lista de secuencias
sequences = [df.iloc[i:i+N] for i in range(0, len(df), N)]

# Visualizar la primera secuencia
sequence = sequences[0]  # Elegir la primera secuencia
for col in columns_of_interest:
    plt.plot(sequence.index, sequence.iloc[:, col], label=f'Feature {col}')

plt.xlabel("Cycle (Index)")
plt.ylabel("Sensor Readings")
plt.title("Sensor Readings Over Cycles for First Sequence")
plt.legend()
plt.show()
```
